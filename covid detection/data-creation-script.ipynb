```python
import os
import pandas as pd
import shutil
import random
```

```python
data = pd.read_csv('metadata.csv')
data.head()
```

```python
covid_data = list(data[(data['finding']=='Pneumonia/Viral/COVID-19') & (data['view']=='PA')]['filename'])
```

```python
for i in range(len(covid_data)):
    shutil.copy2('images/'+covid_data[i],'COVID/'+str(i)+'.jpg')
```

```python
normal_data = os.listdir('../NORMAL')
random.shuffle(normal_data)
for i in range(len(covid_data)):
    shutil.copy2('NORMAL/'+normal_data[i],'norm/'+str(i)+'.jpg')
```
